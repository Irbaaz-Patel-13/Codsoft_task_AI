```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
df = pd.read_csv('netflix_titles_nov_2019.csv')
```

```python
df = df.drop_duplicates(subset='title')
df
```

```python
sample_size = 5000
df = df.sample(n=sample_size, replace=False, random_state=490)
df
```

```python
df = df.reset_index()
df = df.drop(['index', 'show_id'], axis=1)
```

```python
def clean_text(director):
    result = str(director).lower()
    return result.replace(' ', '')
```

```python
df['director'] = df['director'].apply(clean_text)
df['description'] = df['description'].apply(clean_text)
df['title'] = df['title'].str.lower()
df['director'] = df['director'].str.lower()
df
```

```python
df2 = df[['title', 'director', 'description']]
```

```python
df2['data'] = df2[df2.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
```

```python
vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
vectorized
```

```python
similarities = cosine_similarity(vectorized)
df_similarity = pd.DataFrame(similarities, columns=df['title'], index=df['title']).reset_index()
```

```python
input_movie_title = 'the umbrella academy'

recommendations = pd.DataFrame(df_similarity.nlargest(11, input_movie_title)['title'])
recommendations = recommendations[recommendations['title'] != input_movie_title]
print(recommendations)
df
```
